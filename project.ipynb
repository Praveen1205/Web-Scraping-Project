```python
import selenium
import requests
import numpy as np
import pandas as pd
from selenium import webdriver
# from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.maximize_window()
Url = 'https://www.naukri.com/data-science-recruiters-in-delhi-ncr'
driver.get(Url)
```

```python
Name=[]
Designation=[]
Company=[]
Skills=[]
Location=[]

name = driver.find_elements(By.XPATH,'//p[@class="highlightable"]/a[1]/span')
for i in name:
    if i.text is None :
        Name.append("--") 
    else:
        Name.append(i.text)

time.sleep(3)        


designation = driver.find_elements(By.XPATH,'//p[@class="highlightable"]/span[1]')
for i in designation:
    if i.text is None :
        Designation.append("--") 
    else:
        Designation.append(i.text)
        
time.sleep(3)  


company = driver.find_elements(By.XPATH,'//p[@class="highlightable"]/a[2]')
for i in company:
    if i.text is None :
        Company.append("--") 
    else:
        Company.append(i.text)
        
time.sleep(3)  

skills = driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
for i in skills:
    if i.text is None :
        Skills.append("--") 
    else:
        Skills.append(i.text)
        
time.sleep(3)  

location = driver.find_elements(By.XPATH,'//p[@class="highlightable"]/span[2]')
for i in location:
    if i.text is None :
        Location.append("--") 
    else:
        Location.append(i.text)
nxt_button=driver.find_elements(By.XPATH,"//button[@class='grayBtn']")
```

```python
Data = pd.DataFrame({'Name': Name,
                'Designation': Designation,
               'Company': Company,
                'Skills': Skills,
                'Location': Location})
Data
```

```python
Data.to_csv(f"{'Company'}.csv",index=False)
```
